````python
class Node:
    def __init__(self, datav=None):
        self.data = datav
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self , item):
     
        new_node = Node(item)
        
        if not self.head:
            self.head = new_node
        else: 
            current = self.head
            while current.next:
                current = current.next                
            current.next = new_node
    
    def insert(self, index, item):
           
        if index<0:
            raise IndexError("Unsupported Index!")
        
        new_node = Node(item)

        ##empty list
        if not self.head:
            self.head=new_node
            
        ##new head 
        elif index == 0:
            new_node.next =self.head
            self.head = new_node
            
        ##index insertion
        else:
            c = 0
            current =self.head
            prev = None
            
            while current.next and index != c:
                prev = current
                current = current.next
                c += 1
            if index == c:
                prev.next = new_node
                new_node.next = current
            else:
                current.next = new_node
    
            
    def removeNode(self, value):
        current = self.head
        if current.data ==value:
            self.head = current.next
        else:
            prev = None
            while current.next and current.data != value:
                prev = current
                current = current.next
            if current.data == value:
                prev.next = current.next
            else:
                raise ValueError("%s is not in list" % value)
                                    
        

            
````

````python
mylist = LinkedList()

for i in range(5):
    mylist.append(i)
    current = mylist.head


for i in range(5):
    assert current.data==i
    current = current.next
    
mylist.insert(0,-1)
assert mylist.head.data==-1

````
